```python
# ================================================
# Student Grade Calculator
# Author: Godi Anuhya Reddy
# Description: Calculates grades based on marks
# ================================================

def get_grade(marks):
    if marks >= 90:
        return "A+ (Outstanding)"
    elif marks >= 80:
        return "A  (Excellent)"
    elif marks >= 70:
        return "B  (Good)"
    elif marks >= 60:
        return "C  (Average)"
    elif marks >= 50:
        return "D  (Pass)"
    else:
        return "F  (Fail)"

def calculate_percentage(total, max_marks):
    return round((total / max_marks) * 100, 2)

def student_report(name, marks_dict):
    print("=" * 45)
    print(f"       STUDENT REPORT CARD")
    print("=" * 45)
    print(f"  Student Name : {name}")
    print("-" * 45)
    print(f"  {'Subject':<20} {'Marks':>6}  {'Grade'}")
    print("-" * 45)

    total = 0
    max_marks = len(marks_dict) * 100

    for subject, marks in marks_dict.items():
        grade = get_grade(marks)
        print(f"  {subject:<20} {marks:>6}  {grade}")
        total += marks

    percentage = calculate_percentage(total, max_marks)
    overall_grade = get_grade(percentage)

    print("-" * 45)
    print(f"  {'Total Marks':<20} {total:>6}")
    print(f"  {'Percentage':<20} {percentage:>5}%")
    print(f"  {'Overall Grade':<20}       {overall_grade}")
    print("=" * 45)

#Sample Data
student_name = "Godi Anuhya Reddy"

subjects = {
    "Python Programming" : 92,
    "SQL & Databases"    : 88,
    "Data Visualization" : 85,
    "Statistics"         : 79,
    "Communication"      : 95
}

#Run the Report 
student_report(student_name, subjects)
```
